```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import numpy as np
import pandas as pd

vc = pd.read_csv('VistaCustomer.csv')
vc.drop(vc.columns.difference(['EnterpriseId','CustomerId']), 1, inplace=True)
cd = pd.read_csv('CustomerDetail.csv')
merge = vc.merge(cd)

#int: year, string: csvName

def MonthlyByYear(year, csvName):
    
    data=pd.read_csv(csvName)
    data = data.fillna('O')
    data.columns = ['Company', 'Category', 'Cost','Month','Year']
    data=data[data['Year'] == year].reset_index(drop = True)

    subM = pd.DataFrame(data[data['Category'] == 'S'].reset_index(drop = True)
                        .groupby('Month').mean()['Cost'], columns =['Cost'])
    laborM = pd.DataFrame(data[data['Category'] == 'L'].reset_index(drop = True)
                          .groupby('Month').mean()['Cost'], columns =['Cost'])
    materialM = pd.DataFrame(data[data['Category'] == 'M'].reset_index(drop = True)
                             .groupby('Month').mean()['Cost'], columns =['Cost'])
    otherM = pd.DataFrame(data[data['Category'] == 'O'].reset_index(drop = True)
                          .groupby('Month').mean()['Cost'], columns =['Cost'])
    equipmentM = pd.DataFrame(data[data['Category'] == 'E'].reset_index(drop = True)
                              .groupby('Month').mean()['Cost'], columns =['Cost'])
    burdenM = pd.DataFrame(data[data['Category'] == 'B'].reset_index(drop = True)
                           .groupby('Month').mean()['Cost'], columns =['Cost'])
    
    combined = pd.DataFrame({
        'Sub-Contract': subM['Cost'].to_numpy(),
        'Material': materialM['Cost'].to_numpy(),
        'Labor': laborM['Cost'].to_numpy(),
        'Other': otherM['Cost'].to_numpy(),
        'Equipment': equipmentM['Cost'].to_numpy(),
        'Burden': burdenM['Cost'].to_numpy(),
        },
        index=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    )
    
    return combined

#int: year, string: vert, string: csvName

def VertMonthlyByYear(year, vert, csvName):
    
    vc = pd.read_csv('VistaCustomer.csv')
    vc.drop(vc.columns.difference(['EnterpriseId','CustomerId']), 1, inplace=True)
    cd = pd.read_csv('CustomerDetail.csv')
    merge = vc.merge(cd)
    merge.rename(columns={'EnterpriseId':'Company'}, inplace=True)
    merge.drop(merge.columns.difference(['Company', 'Vertical']), 1, inplace = True)
    
    data=pd.read_csv(csvName)
    data = data.fillna('O')
    data.columns = ['Company', 'Category', 'Cost','Month','Year']
    data = data.merge(merge, on = 'Company')
    data= data[data['Year'] == year].reset_index(drop = True)
    data = data[data['Vertical'] == vert]

    subM = pd.DataFrame(data[data['Category'] == 'S'].reset_index(drop = True)
                        .groupby('Month').mean()['Cost'], columns =['Cost'])
    laborM = pd.DataFrame(data[data['Category'] == 'L'].reset_index(drop = True)
                          .groupby('Month').mean()['Cost'], columns =['Cost'])
    materialM = pd.DataFrame(data[data['Category'] == 'M'].reset_index(drop = True)
                             .groupby('Month').mean()['Cost'], columns =['Cost'])
    otherM = pd.DataFrame(data[data['Category'] == 'O'].reset_index(drop = True)
                          .groupby('Month').mean()['Cost'], columns =['Cost'])
    equipmentM = pd.DataFrame(data[data['Category'] == 'E'].reset_index(drop = True)
                              .groupby('Month').mean()['Cost'], columns =['Cost'])
    burdenM = pd.DataFrame(data[data['Category'] == 'B'].reset_index(drop = True)
                           .groupby('Month').mean()['Cost'], columns =['Cost'])
    
    combined = pd.DataFrame({
        'Sub-Contract': subM['Cost'].to_numpy(),
        'Material': materialM['Cost'].to_numpy(),
        'Labor': laborM['Cost'].to_numpy(),
        'Other': otherM['Cost'].to_numpy(),
        'Equipment': equipmentM['Cost'].to_numpy(),
        'Burden': burdenM['Cost'].to_numpy(),
        },
        index=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    )
    
    return combined

#int: year, string: csvName

def YearlyStats(year, csvName):
    
    data=pd.read_csv(csvName)
    data = data.fillna('O')
    data.columns = ['Company', 'Category', 'Cost','Month','Year']
    data=data[data['Year'] == year].reset_index(drop = True)

    subM = data[data['Category'] == 'S'].sum()['Cost']
    laborM = data[data['Category'] == 'L'].sum()['Cost']
    materialM = data[data['Category'] == 'M'].sum()['Cost']
    equipmentM = data[data['Category'] == 'E'].sum()['Cost']
    otherM = data[data['Category'] == 'O'].sum()['Cost']
    burdenM = data[data['Category'] == 'B'].sum()['Cost']

    combined = pd.DataFrame({
        'Sub-Contract': subM,
        'Material': materialM,
        'Labor': laborM,
        'Other': otherM,
        'Equipment': equipmentM,
        'Burden': burdenM,
        }, 
        index = [year]     
    )
    return combined

#int: year, string: vert, string: csvName

def VertYearlyStats(year, vert, csvName):
    
    vc = pd.read_csv('VistaCustomer.csv')
    vc.drop(vc.columns.difference(['EnterpriseId','CustomerId']), 1, inplace=True)
    cd = pd.read_csv('CustomerDetail.csv')
    merge = vc.merge(cd)
    merge.rename(columns={'EnterpriseId':'Company'}, inplace=True)
    merge.drop(merge.columns.difference(['Company', 'Vertical']), 1, inplace = True)
    
    data=pd.read_csv(csvName)
    data = data.fillna('O')
    data.columns = ['Company', 'Category', 'Cost','Month','Year']
    data = data.merge(merge, on = 'Company')
    data= data[data['Year'] == year].reset_index(drop = True)
    data = data[data['Vertical'] == vert]

    subM = data[data['Category'] == 'S'].sum()['Cost']
    laborM = data[data['Category'] == 'L'].sum()['Cost']
    materialM = data[data['Category'] == 'M'].sum()['Cost']
    equipmentM = data[data['Category'] == 'E'].sum()['Cost']
    otherM = data[data['Category'] == 'O'].sum()['Cost']
    burdenM = data[data['Category'] == 'B'].sum()['Cost']

    combined = pd.DataFrame({
        'Sub-Contract': subM,
        'Material': materialM,
        'Labor': laborM,
        'Other': otherM,
        'Equipment': equipmentM,
        'Burden': burdenM,
        }, 
        index = [year]     
    )
    return combined

#DataFrame: data, string: pltName, Returns a stacked bar Plot

def StackedBar (data, pltName):
    #yt = [0,1000000,2000000,3000000,4000000,5000000,6000000,7000000,8000000,9000000,10000000]
    myplt = data.plot(kind='bar', stacked=True, figsize=(16,8), colormap = 'Blues_r', title = pltName, rot = 1, fontsize = 12)
    myplt.set_facecolor('lightgray')
    myplt.legend(loc='lower left', bbox_to_anchor=(1.0, 0.5))
    save = myplt.get_figure()
    save.savefig(pltName + '.png')
```
